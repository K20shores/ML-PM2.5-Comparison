```python
import pandas as pd
```

```python
df = pd.read_csv('data/matched/all_years_matched.csv', parse_dates=['time'])
```

```python
drop = ['Local Site Name',
       'Address', 'State Name', 'County Name', 'City Name', 'proj_longitude',
       'proj_latitude', 'index', 'POC']
df = df.drop(columns=drop, axis=1)
```

```python
df
```

```python
seasons_df = df.copy()
```

```python
months_df = df.copy()
```

# Making seasons

```python
sp = 'spring'
f = 'fall'
sm = 'summer'
w = 'winter'
seasons = [w, w, sp, sp, sp, sm, sm, sm, f, f, f, w]

month_to_season = dict(zip(range(1,13), seasons))
seasons_df['season'] = seasons_df.time.dt.month.map(month_to_season)
seasons_df.season = seasons_df.season.astype('category')
```

```python
seasons_df = pd.get_dummies(seasons_df[seasons_df.columns])
```

```python
seasons_df
```

```python
seasons_df.to_csv('data/matched/all_years_final_seasons.csv', index=False)
```

# Making Months

```python
months_df['month'] = months_df.time.dt.month
months_df.month = months_df.month.astype('category')
```

```python
months_df = pd.get_dummies(months_df[months_df.columns])
```

```python
months_df
```

```python
months_df.to_csv('data/matched/all_years_final_months.csv', index=False)
```
