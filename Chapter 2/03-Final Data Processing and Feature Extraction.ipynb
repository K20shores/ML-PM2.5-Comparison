```python
import pandas as pd
import glob
```

```python
df = pd.concat([
    pd.read_csv(f, parse_dates=['time']) for f in sorted(glob.glob('/Volumes/Canon/matched/2021/*.csv'))
], axis=0)
```

```python
df
```

```python
df.dropna()
```

```python
drop = ['AE', 'AOD470']
df = df.drop(columns=drop, axis=1)
```

```python
df
```

```python
seasons_df = df.copy()
```

```python
months_df = df.copy()
```

# Making seasons

```python
sp = 'spring'
f = 'fall'
sm = 'summer'
w = 'winter'
seasons = [w, w, sp, sp, sp, sm, sm, sm, f, f, f, w]

month_to_season = dict(zip(range(1,13), seasons))
seasons_df['season'] = seasons_df.time.dt.month.map(month_to_season)
seasons_df.season = seasons_df.season.astype('category')
```

```python
seasons_df = pd.get_dummies(seasons_df[seasons_df.columns])
```

```python
seasons_df
```

```python
seasons_df.to_csv('data/matched/2021_seasons.csv', index=False)
```

# Making Months

```python
months_df['month'] = months_df.time.dt.month
months_df.month = months_df.month.astype('category')
```

```python
months_df = pd.get_dummies(months_df[months_df.columns])
```

```python
months_df
```

```python
months_df.to_csv('data/matched/2021_months.csv', index=False)
```
