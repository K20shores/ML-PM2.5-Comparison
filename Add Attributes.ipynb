```python
import xarray as xr
import os
import glob
```

```python
products = ['M2I1NXASM','M2I3NXGAS','M2SDNXSLV','M2T1NXAER','M2T1NXRAD']
```

```python
files = [glob.glob(
        os.path.join('data', 'MERRA-2', product, '*.nc4')
    )[0] for product in products]
```

```python
files
```

```python
attrs = {}

for file in files:
    with xr.open_dataset(file) as ds:
        ds.load()
    for var in ds.data_vars:
        attrs[var] = ds[var].attrs
```

```python
attrs['U2M']
```

```python
path = 'data/MERRA-2/Texas/merged-all-products-daily-averages-clipped.nc4'
```

```python
with xr.open_dataset(path) as ds:
        ds.load()
ds
```

```python
for var, attr in attrs.items():
    ds[var].attrs = attr
```
