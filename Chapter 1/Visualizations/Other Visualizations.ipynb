```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('/Users/kyle/GoogleDrive/personal.mplstyle')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=20, n_features=1, random_state=0, noise=4.0,
                       bias=100.0)

reg = LinearRegression().fit(X, y)

fig, ax = plt.subplots(dpi = 300)

line = X * reg.coef_[0] + reg.intercept_
ax.plot(X, line, color='red', alpha=0.5)

ax.scatter(X, y, color='blue', s=50, alpha=0.4, edgecolors='black', linewidth=.5, zorder=3)

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.tick_params(axis='both', labelsize=6, width=0)
ax.set_xticklabels([])
ax.set_yticklabels([])

fig.savefig('regresion-line.png', format='png')
```
