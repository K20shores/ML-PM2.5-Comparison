```python
import geopandas as gpd
import xarray as xr
import os
import numpy as np
import matplotlib.animation as anim
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
ds = xr.open_dataset(os.path.join('data', 'MERRA-2', 'Texas', 'merged-all-products-daily-averages.nc4'))
ds
```

```python
gdf = gpd.read_file(os.path.join('data', 'Texas_State_Boundary-shp', 'State.shp'))
gdf
```

```python
ds.rio.set_spatial_dims('lon', 'lat', inplace=True)
ds.rio.write_crs('EPSG:4326', inplace=True)

clipped = ds.rio.clip(
    [gdf.__geo_interface__['features'][0]['geometry']], 
    crs='EPSG:4326'
)
clipped
```

```python
ds.close()
```

```python
clipped.to_netcdf(os.path.join('data', 'MERRA-2', 'Texas', 'merged-all-products-daily-averages-clipped.nc4'))
```
