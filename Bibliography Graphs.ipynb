```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import json

# read file
with open('bibliography.json', 'r') as myfile:
    data=myfile.read()
bib = json.loads(data)
```

```python
models = set([b for a in bib for b in a['models']])
```

```python
years = [item['year'] for item in bib]
```

```python
model_use_by_year = {year: {model: 0 for model in models} for year in range(min(years), max(years)+1)}
```

```python
for item in bib:
    for model in item['models']:
        model_use_by_year[item['year']][model] += 1
```

```python
model_use_by_year
```

```python
df = pd.DataFrame(model_use_by_year).T
```

```python
fig, ax = plt.subplots(dpi=300)

df.plot.bar(legend = False, ax=ax)
```
